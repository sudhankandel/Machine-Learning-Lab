```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import math
from math import exp
```

```python
df=pd.read_csv('Diabetes.csv')
independent_variable=df[['Pragnency', 'Glucose', 'Blod Pressure', 'Skin Thikness', 'Insulin',
       'BMI', 'DFP', 'Age']]
target=df['Diabetes']
x_train,x_test,y_train,y_test=train_test_split(independent_variable,target,train_size=0.7)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
predicted_value=model.predict(x_test)
```

```python
print(metrics.classification_report(predicted_value, y_test))
```

```python
x_train['target']=y_train
```

```python
def coffiecient_bgd(train,l_rate,n_epoch):
    train=train.to_numpy()
    
    coef=[0.0 for i in range(len(train[0]))]
    for j in range(n_epoch):
        print("..............epoch"+str(j)+"...............")
        for i in train:
            z=coef[0]+coef[1]*i[0]+coef[2]*i[1]+coef[3]*i[2]+coef[4]*i[3]+coef[5]*i[4]+coef[6]*i[5]+coef[7]*i[6]+coef[8]*i[7]
            yhat=1/(1+exp(-z))
            coef[0]-=l_rate*(yhat-i[8])
            coef[1]-=l_rate*(yhat-i[8])*i[0]
            coef[2]-=l_rate*(yhat-i[8])*i[1]
            coef[3]-=l_rate*(yhat-i[8])*i[2]
            coef[4]-=l_rate*(yhat-i[8])*i[3]
            coef[5]-=l_rate*(yhat-i[8])*i[4]
            coef[6]-=l_rate*(yhat-i[8])*i[5]
            coef[7]-=l_rate*(yhat-i[8])*i[6]
            coef[8]-=l_rate*(yhat-i[8])*i[7]
            print("Updated Weights values are ",coef)
    return coef
```

```python

coef=coffiecient_bgd(x_train,0.001,30)
```

```python
def prediction(coef,test):
    test=test.to_numpy()
    predict=[]
    for i in test:
        y=coef[0]+coef[1]*i[0]+coef[2]*i[1]+coef[3]*i[2]+coef[4]*i[3]+coef[5]*i[4]+coef[6]*i[5]+coef[7]*i[6]+coef[8]*i[7]
        yhat=1/(1+exp(-y))
        if yhat>0.5:
            predict.append(1)
        else:
            predict.append(0)
    return predict
```

```python
predicted_value=prediction(coef,x_test)
```

```python
print(metrics.classification_report(predicted_value, y_test))
```
